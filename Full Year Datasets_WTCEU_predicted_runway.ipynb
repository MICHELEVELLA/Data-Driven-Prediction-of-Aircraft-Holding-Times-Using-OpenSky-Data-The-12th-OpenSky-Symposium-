```python
import requests
import json

doc8643_url = r = "https://www4.icao.int/doc8643/External/AircraftTypes"
r = requests.post(doc8643_url)
doc8643at = pd.json_normalize(r.json())
doc8643at.columns = doc8643at.columns.str.lower()
doc8643at = doc8643at.rename(columns={
    "modelfullname": "model",
    "manufacturercode": "manufacturer_code",
    "aircraftdescription": "aircraft_description",
    "enginecount": "engine_count",
    "enginetype": "engine_type",
    "wtg":"RECAT_EU"
    })

def Icao_doc8643(df):
    for index,row in df.iterrows():
        type_code=None
        RECAT_EU=None
        type_code=row['Type Code']

        if pd.notna(row['Type Code']):
            icao_row = doc8643at[doc8643at['designator'] == type_code].iloc[0]
            RECAT_EU=icao_row["RECAT_EU"]

        if RECAT_EU=='A':
            df.at[index,'RECAT_EU_A']=1
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=0


        elif RECAT_EU=='B':
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=1
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=0


        elif RECAT_EU=='C':
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=1
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=0


        elif RECAT_EU=='D':
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=1
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=0

        elif RECAT_EU=='E':
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=1
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=0

        elif RECAT_EU=='F':
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=1
            df.at[index,'RECAT_EU_G']=0

        elif RECAT_EU=='G':
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=1

        elif RECAT_EU=='Z':
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=0

        elif RECAT_EU==None:
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=0
        else:
            df.at[index,'RECAT_EU_A']=0
            df.at[index,'RECAT_EU_B']=0
            df.at[index,'RECAT_EU_C']=0
            df.at[index,'RECAT_EU_D']=0
            df.at[index,'RECAT_EU_E']=0
            df.at[index,'RECAT_EU_F']=0
            df.at[index,'RECAT_EU_G']=0
        
    return df
```

```python
import pandas as pd


# 1) Create a reference DataFrame of TimeStamp + Runway columns, rename TimeStamp -> ref_time
runway_df = (
    Df_w_RECAT_EU[['Last Timestamp', 'Runway 09L', 'Runway 09R', 'Runway 27L', 'Runway 27R']]
    #.rename(columns={'TimeStamp': 'ref_time'})
    .sort_values('Last Timestamp')
)

runway_df['Last Timestamp'] = pd.to_datetime(
    runway_df['Last Timestamp'], 
    format='%Y-%m-%d %H:%M:%S'
)

Df_w_RECAT_EU['TimeStamp'] = pd.to_datetime(
    Df_w_RECAT_EU['TimeStamp'], 
    format='%Y-%m-%d %H:%M:%S'
)

Df_w_RECAT_EU['TimeStamp_minus_30min'] = Df_w_RECAT_EU['TimeStamp'] - pd.Timedelta(minutes=30)
Df_w_RECAT_EU['TimeStamp_minus_60min'] = Df_w_RECAT_EU['TimeStamp'] - pd.Timedelta(minutes=60)

#runway_df['ref_time_minus_60min'] = runway_df['ref_time'] - pd.Timedelta(minutes=60)


# 2) Sort your main DataFrame by the 'Last Timestamp' for merge_asof
#main_df = Df_w_RECAT_EU.sort_values('Last Timestamp').copy()
```

```python
# 3) Merge asof: left_on = 'Last Timestamp', right_on = 'ref_time'
#    direction='nearest' ensures we pick the closest match in time
Df_w_RECAT_EU=Df_w_RECAT_EU.sort_values('TimeStamp_minus_30min')

merged_df = pd.merge_asof(
    Df_w_RECAT_EU, 
    runway_df, 
    left_on='TimeStamp_minus_30min', 
    right_on='Last Timestamp',
    direction='nearest',
    suffixes=('', ' predicted at -30')  # We'll keep a suffix for the 'ref_time' DataFrame
)


# Now merged_df contains the original columns plus the four newly added "Actual Runway ..." columns.
```
